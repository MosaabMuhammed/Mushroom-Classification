**looks like our data has a column with only one unique value "veil-type", since it has one value, I'm going to remove it.**

```python
del df_copy['veil-type']
```

```python
df_copy.columns
```

**Now, since the 21 feature columns are labeled with letters, I'm going to convert them to integers**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in df_copy.columns:
    df_copy[col] = le.fit_transform(df_copy[col])

df_copy.sample(8)
```

**Since, we have 21 features, we need to reduce the dimensions by using PCA**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(df_copy.iloc[:, 1:21])
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit_transform(df_copy.iloc[:, 1:21])
```

```python
df_copy.columns
```

```python
explained_vairance = pca.explained_variance_
```

```python
with plt.style.context('dark_background'):
    plt.figure(figsize=(6, 4))
    plt.bar(range(21), explained_vairance, alpha=.5, align='center',
           label='individula explained vari')
```
